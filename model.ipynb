```python
import pandas as pd
df = pd.read_csv("../data/customer_support_tickets.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
import nltk, re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = word_tokenize(text)
    return " ".join([w for w in words if w not in stop_words])

df['cleaned'] = df['ticket'].apply(preprocess)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['cleaned'])
```

```python
df[['ticket', 'cleaned']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['cleaned'])
```

```python
from sklearn.linear_model import LogisticRegression

model_cat = LogisticRegression(max_iter=1000)
model_cat.fit(X, df['category'])

model_pri = LogisticRegression(max_iter=1000)
model_pri.fit(X, df['priority'])
```

```python
import joblib

joblib.dump(model_cat, "../model/category_model.pkl")
joblib.dump(model_pri, "../model/priority_model.pkl")
joblib.dump(vectorizer, "../model/vectorizer.pkl")
```
